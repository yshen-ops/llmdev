```python
# 导入所需模块
import  os 
from  dotenv  import  load_dotenv 
from  openai  import  OpenAI

# 获取环境变量
load_dotenv( "../.env " )

# 创建一个 OpenAI API 客户端
client = OpenAI(api_key=os.environ['API_KEY'])

# 获取模型列表
models = client.models.list()

# 列出models中的每个model的模型：
for model in models.data:
    print(model.id)
```
