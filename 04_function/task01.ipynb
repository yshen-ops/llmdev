```python
def check(str):
    if len(str) < 5 or len(str) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    if str == "" or " " in str:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    if not str.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    return True

def output(str):
    if check(str):
        print("ユーザー名は有効です")
    else:
        print("不正なユーザー名です")
        
output("JohnDoe123")
output("JD")
output("John@Doe")
output("John Doe")
```
