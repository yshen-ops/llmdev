```python
# 数値を格納したリスト
numbers = [1, 2, 3, 4, 5]
print(numbers) # [1, 2, 3, 4, 5]

# 文字列を格納したリスト
fruits = ["apple", "banana", "cherry"]
print(fruits) # ['apple', 'banana', 'cherry']

# 異なるデータ型を混在させたリスト
mixed = [1, "apple", 3.14, True]
print(mixed) # [1, 'apple', 3.14, True]

# 空のリスト
empty_list = []
print(empty_list) # []

# 元のリスト
original = [1, 2, 3]

# スライスによるコピー
copy1 = original[:]
print(copy1)  # [1, 2, 3]

# list() 関数によるコピー
copy2 = list(original)
print(copy2)  # [1, 2, 3]

# copy() メソッドによるコピー
copy3 = original.copy()
print(copy3)  # [1, 2, 3]

import copy

# 元のリスト（多次元リスト）
original_list = [[1, 2], [3, 4]]

# 浅いコピーを作成
shallow_copied_list = copy.copy(original_list)
# 深いコピーを作成
deep_copied_list = copy.deepcopy(original_list)

# 浅いコピーのリストを変更
shallow_copied_list[0][0] = 100
# 深いコピーのリストを変更
deep_copied_list[1][1] = 200

# 結果の出力
print("元のリスト:", original_list)             
print("浅いコピー:", shallow_copied_list)       
# → 浅いコピーの変更が元のリストに影響している

print("深いコピー:", deep_copied_list)         
# → 深いコピーの変更は元のリストには影響しない


fruits = ["apple", "banana", "cherry", "banana"]

# 指定インデックスの要素を削除（インデックス1）
del_item = fruits.pop(1)
print(fruits)  # ["apple", "cherry", "banana"]
print(del_item) # banana

# 指定の値に一致する最初の要素を削除
fruits.remove("banana")
print(fruits)  # ["apple", "cherry"]

# del キーワードで指定の要素を削除（インデックス0）
del fruits[0]
print(fruits)  # ["cherry"]

# 全要素を削除して空リストにする
fruits.clear()
print(fruits)  # []

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# 2次元リストをネストしたループで表示
for row in matrix:
    for element in row:
        print(f"{element:1d}", end="") # 改行を出力しない指定
    print()  # 改行
```
