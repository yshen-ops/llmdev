```python
import string

words = []
count = {char: 0 for char in string.ascii_lowercase}
while True:
    word = input("英単語を入力してください：")
    if word == "":
        break
    words.append(word)
    for i in word:
        count[i] += 1        

words.sort()
print(f"入力した英単語：{words}")
for key, value in count.items():
    print(f"{key}が{value}個ありました")
```
