```python
# 数値を格納した辞書
numbers = {"one": 1, "two": 2, "three": 3}
print(numbers)  # {'one': 1, 'two': 2, 'three': 3}

# 異なるデータ型の値を格納した辞書
person = {"name": "Alice", "age": 25, "is_student": False}
print(person)  # {'name': 'Alice', 'age': 25, 'is_student': False}

# 空の辞書
empty_dict = {}
print(empty_dict)  # {}
```
