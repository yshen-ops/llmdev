```python
import requests
import json

def tokyo_week_raw():
    url = "https://www.jma.go.jp/bosai/forecast/data/forecast/130000.json"
    data = requests.get(url, timeout=10).json()

    return data[1]

data = tokyo_week_raw()
```

```python
# 必要なモジュールをインポート
import os
from dotenv import load_dotenv
from openai import OpenAI

# 環境変数の取得
load_dotenv("../.env")

# OpenAI APIクライアントを生成
client = OpenAI(api_key=os.environ['API_KEY'])

# モデル名
MODEL_NAME = "gpt-4o-mini"

# 要約を行うプロンプトを作成
prompt = f"""
以下は東京の週間天気データです。
今週の外出アドバイスを簡潔にまとめてください。

# データ：
{data}
"""

# APIへリクエスト
response = client.chat.completions.create(
    model=MODEL_NAME,
    messages=[
        {"role": "user", "content": prompt},
    ],
    max_tokens=500,
    temperature=0.3
)

```
